```python
import io
import torch
from torchtext.utils import download_from_url, extract_archive
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
```

```python
tokenizer = get_tokenizer('basic_english')
```

```python
%%time
sample_texts = bc_dataset['train']['text']

# 9gb in memory
```

```python
vocab = load_vocab('bookcorpus-vocab.pkl')
```

```python
%%time
def data_process(raw_text_iter):
    data = [torch.tensor([vocab[token] for token in tokenizer(item)], dtype=torch.long) for item in raw_text_iter]
    return torch.cat(tuple(filter(lambda t: t.numel() > 0, data)))


train_data = data_process(iter(sample_texts))
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
vocab = build_vocab_from_iterator(map(tokenizer,iter(sample_texts)))
```

```python
def save_vocab(vocab, path):
    import pickle
    output = open(path, 'wb')
    pickle.dump(vocab, output)
    output.close()
    
def load_vocab(path):
    import pickle
    output = open(path, 'rb')
    vocab = pickle.load(output)
    output.close()
    return vocab
```

```python
save_vocab(vocab, 'bookcorpus-vocab.pkl')
```
